```python
import h5py
import random
import numpy as np
import utils_visu
from matplotlib import pyplot as plt
import plotly.offline as py
py.init_notebook_mode(connected=True)
%load_ext autoreload
%autoreload 2
```

```python
PATH_DATA = 'data/train/eightieth.h5'
```

```python
data = h5py.File(PATH_DATA,'r')
```

```python
nb_patches = data["S2"].shape[0]
sample_list = np.random.choice(np.arange(nb_patches), 100, replace=False)
sample_list = np.sort(sample_list)
sample_list = [4]
```

```python
np.set_printoptions(threshold=np.nan)
original = data["S2"][sample_list,:,:,0:3]
orig = original * 255 / 1400
orig = orig.astype('uint8')
cdf = utils_visu.get_equalization_table(original)
res = cdf[original.astype('uint16')]
```
